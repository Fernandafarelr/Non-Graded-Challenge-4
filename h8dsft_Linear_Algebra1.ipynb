```python
import numpy as np

v = np.array([17, 22, 19])

print(v)
```

```python
A = np.array([17,22,19])
B = np.array([10,20,11])
C = np.array([5,12,9])

addition = A + B
print("A + B = ", addition)

subtraction = B - C
print("B - C = ", subtraction)

dot_product = np.dot(A, C)
print("A dot C = ", dot_product)

cross_product = np.cross(A, B)
print("A x B = ", cross_product)

norm_A = np.linalg.norm(A)
print("||A|| = ", norm_A)

cosine_similarity = np.dot(A, B) / (np.linalg.norm(A) * np.linalg.norm(B))
angle = np.arccos(cosine_similarity)
print("angle between A and B (in radians) =", angle)
```

```python
import matplotlib.pyplot as plt

# Define the vectors u and v
u = [2, 5]
v = [3, 1]

# Create a new figure and set the axis limits
fig, ax = plt.subplots()
ax.set_xlim([0, 6])
ax.set_ylim([0, 6])

# Plot the vectors u and v as arrows
ax.arrow(0, 0, u[0], u[1], head_width=0.3, head_length=0.5, fc='blue', ec='blue')
ax.arrow(0, 0, v[0], v[1], head_width=0.3, head_length=0.5, fc='green', ec='green')

# Add labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Vectors u = (2, 5) and v = (3, 1)')

# Show the plot
plt.show()
```
